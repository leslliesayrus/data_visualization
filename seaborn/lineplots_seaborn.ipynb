```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet('part-00000-dac123ca-8acd-4f0c-b541-f5a5ed4bf12c-c000.snappy (1).parquet', engine='pyarrow')
```

```python
df["date"] = pd.to_datetime(df["date"])
df = df.set_index('date')
```

```python
df
```

```python
sns.lineplot(data = df['close'])
```

```python
# size of image
plt.figure(figsize=(14,6))

# Add title
plt.title("Bitcoin Price")

# The graph
sns.lineplot(data = df[['close', 'high']])
plt.show()
```

```python
# multiple plots
# size of image
plt.figure(figsize=(14,6))
# Add title
plt.title("Bitcoin Price")
# The graph
sns.lineplot(data = df['low'], label = 'low')
sns.lineplot(data = df['high'], label = 'high')
plt.show()
```
