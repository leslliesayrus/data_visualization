```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet('part-00000-dac123ca-8acd-4f0c-b541-f5a5ed4bf12c-c000.snappy (1).parquet', engine='pyarrow')
```

```python
df
```

```python
month = [int(df.iloc[i,0][1]) for i in range(len(df))]
```

```python
df['month'] = month
```

```python
df.head(2)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))
# Add title
plt.title("variability per month of Bitcoin")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=df['month'], y=df['variability'])

# Add label for vertical axis
plt.ylabel("variability")
```
